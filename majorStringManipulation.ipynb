```python
name = "Muskan"        # store name as string
age = 21               # store age as integer
marks = 87.45          # store marks as float

print("Name: %s" % name)        # print string using %s
print("Age: %d" % age)          # print integer using %d
print("Marks: %.2f" % marks)    # print float with 2 decimal places

print("Name width 10: %10s" % name)   # print name with space before
print("Age zero padded: %05d" % age)  # print age with leading zeros

print("Student %s scored %.1f marks" % (name, marks))  
# print multiple values together
```

```python
# Example 1: Mobile Recharge Details
mobile = "9876543210"       # store mobile number
plan = "Unlimited"         # store plan type
amount = 299.99            # recharge amount

print("Mobile Number: %s" % mobile)   # print mobile number
print("Plan Type: %s" % plan)         # print plan
print("Recharge Amount: %.2f" % amount)  
# print amount with 2 decimal places
```

```python
# Example 2: Movie Ticket Booking
movie = "Avengers"          # store movie name
tickets = 3                # store number of tickets
price = 250.0              # price per ticket

print("Movie Name: %s" % movie)          # print movie name
print("Tickets: %d" % tickets)           # print number of tickets
print("Price per Ticket: %.2f" % price)  # print ticket price
print("Total Cost: %.2f" % (tickets * price))  
```

```python
# Example 3: Grocery Item Bill Calculation
item = "Rice"           # store item name
weight = 5              # store weight in kg
rate = 60.75            # store rate per kg

print("Item Name: %s" % item)               # print item name
print("Weight: %d kg" % weight)             # print weight
print("Rate per kg: %.2f" % rate)           # print rate
print("Total Amount: %.2f" % (weight * rate))  
# calculate and print total amount
```

```python
# Example 4: Food Order Bill Calculation
food = "Pizza"          # store food name
quantity = 2            # store quantity
price = 150.5           # store price of one item

print("Food Item: %s" % food)              # print food name
print("Quantity: %d" % quantity)           # print quantity
print("Price per item: %.2f" % price)      # print price with 2 decimals
print("Total Bill: %.2f" % (quantity * price))  
# calculate and print total bill
```
