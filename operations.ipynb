```python
#Arithmetic operators
# +, -, /, //, %, **

#Addition example
print("Addition of two numbers")
num1 = int(input("Enter First Number"))
num2 = int(input("Enter second number"))
add = num1+num2
print("Addition is:", add)

#Subtraction example
print("Substraction of two numbers")
num1 = int(input("Enter First Number"))
num2 = int(input("Enter second number"))
sub = num1 - num2
print("Substraction is:", sub)

#concatination example
print("Concatination two numbers")
sen_1= input("Enter the first number")
sen_2= input("Enter the second number")
print(sen_1+sen_2)

#(/) use for division
print("Division two numbers")
num1 = int(input("Enter first number:"))
num2 = int(input("Enter first number:"))
div = num1/num2
print("Division is:", div)

#(%)use for modules
print("Modulus two number")
num3 = int(input("Enter first number:"))
num4 = int(input("Enter first number:"))
mod = num2 % num4
print("modulus is: ", mod)

#(//) use for floor division
print("Floor divion two numbers")
num5 = int(input("Enter first number:"))
num6 = int(input("Enter first number:"))
floordiv = num4 // num6
print("Floor division is :", floordiv)

#(**) use for exponentiation
print("Exponentiation of two numbers")
num7 = int(input("Enter first number:"))
num8 = int(input("Enter first number:"))
exp = num7 ** num8
print("Exponentiation is :", exp)
```
